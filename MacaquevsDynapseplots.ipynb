```python
from scipy.optimize import fsolve
def findIntersection(fun1,fun2,x0):
    return fsolve(lambda x : fun1(x) - fun2(x),x0)
```

```python
ax, fig1 = plt.subplots()
plt.loglog((temporal_expand[:]),RInhRmean[:, 4]*50/np.max(RRdsmean),'-o', color = "#99d6ff", label = 'S')
plt.loglog((temporal_expand[:]),RRdsmean[:, 4]*50/np.max(RRdsmean),'-s',color = "#66ffc2", label = 'T')
plt.loglog(xp*50,p(xp)*50/np.max(p(xp)), color = "#00995c")
plt.loglog(hm*50,m(hm)*50/np.max(p(xp)), color = "#008ae6")
fig1.set_ylabel('Normalized Response',fontsize=20)
fig1.set_xlabel('Temporal Frequency (cycles/s)', fontsize=20)
fig1.legend(loc='upper right', title="sf = 0.0621c/deg",fontsize = 13,frameon = False)
ax.set_size_inches(6, 6)
plt.axvline(1.12129984, color="#c2d6d6", linestyle='--')
fig1.set_xlim([10**1, 20**2])
fig1.spines['top'].set_visible(False)
fig1.spines['right'].set_visible(False)
plt.show()

print findIntersection(m,p,0.0)
```
